```python
import requests
from bs4 import BeautifulSoup
from PIL import Image
from io import BytesIO

def get_bird_info(bird_name):
    url = f'https://en.wikipedia.org/wiki/{bird_name}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the image element using XPath
    img_element = soup.find('img', {'class': 'thumbimage'})
    img_src = img_element['src'] if img_element else None
    
    # Get the image
    if img_src:
        img_response = requests.get(f'https:{img_src}')
        img = Image.open(BytesIO(img_response.content))
        img.show()  # Display the image using PIL
    
    # Find and print other information about the bird
    # Example: Scientific classification
    classification = soup.find('table', {'class': 'infobox'}).find('tr', {'class': 'biota'}).text.strip()
    print('Bird Name:', bird_name.replace('_', ' '))
    print('Scientific Classification:', classification)

# Example usage
get_bird_info('African_grey_parrot')
```

```python
import requests
import re
from bs4 import BeautifulSoup

def get_image_src_links(breedname):
    wikipedia_url = f'https://en.wikipedia.org/wiki/{breedname}'

    response = requests.get(wikipedia_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all img elements with class mw-file-element
    image_elements = soup.find_all('img', {'class': 'mw-file-element'})
    
    # Extract src links from img elements
    src_links = [img['src'] for img in image_elements]
    links = []
    for num, link in enumerate(src_links):
        if "svg" in links:
            print("svg present")
        else:
            pattern = r'/(\d+)px'
        # Define the replacement string
            replacement = r'/1000px'

        # Use re.sub to replace matching patterns
            modified_link = re.sub(pattern, replacement,link)
            if (num < 3):
                links.append(modified_link)
    return links
```
